# Executive Margin Protection Summary

## Business Problem

Revenue is growing, but contribution margin is shrinking.

Leadership needed to understand which customer segments, pricing decisions, and regional operations were driving profitability erosion.

The goal was to identify where revenue growth was becoming financially destructive and recommend actions to protect margin without slowing growth.

---

## Key Findings

### 1. Promo Hunters Created Negative Growth

Promo Hunters generated over $232K in revenue but created more than $80K in losses.

This segment showed:

* 32.7% average discounting
* lowest retention (0.23)
* highest refund abuse
* highest CAC inefficiency

This indicated top-line growth was being purchased through unsustainable promotional dependency rather than profitable customer acquisition.

---

### 2. Discounts Above 20% Destroyed Contribution Margin

Discount levels above 20% consistently produced negative margins.

At 30%+ discount levels:

## Contribution Margin = -64.6%

This created a clear financial threshold where discounting shifted from customer acquisition strategy to margin destruction.

The analysis showed that top-line growth was being purchased at the expense of profitability.

---

### 3. West Region Was Highest Revenue but Negative Profit

The West region generated the highest revenue but operated at negative contribution profit.

Primary drivers included:

* highest refund rate
* highest driver incentive costs
* operational inefficiency

This showed that revenue concentration was masking profitability risk and operational leakage.

---

### 4. Corporate Lunch + Subscribers Were Highest-Value Segments

Corporate Lunch delivered the strongest contribution margin with minimal discount dependency.

Subscribers also showed:

* stronger margins
* lower refund rates
* better retention
* lower margin volatility

These segments became the strongest targets for CAC reallocation and sustainable growth.

---

## Executive Recommendations

### Recommendation 1

Cap discounts above 20%, especially for low-retention promo-heavy cohorts.

This protects contribution margin while reducing unprofitable promotional dependency.

---

### Recommendation 2

Reallocate CAC away from Promo Hunters toward:

* Corporate Lunch
* Loyal Subscribers

This improves acquisition efficiency and increases long-term customer value.

---

### Recommendation 3

Expand subscription strategy to improve retention and long-term contribution margin.

DashPass-style behavior creates stronger profitability and reduces dependency on short-term promotions.

---

### Recommendation 4

Reduce operational leakage in the West region through:

* refund reduction initiatives
* delivery quality improvement
* driver cost optimization

This turns the highest-revenue region into a profitable growth engine.

---

## Expected Business Impact

These changes are expected to:

* recover contribution margin
* improve acquisition efficiency
* increase customer lifetime value
* reduce operational leakage
* shift growth toward profitable customer segments

The goal is not simply more revenue.

The goal is better revenue.

Profitable growth is the strategy.
