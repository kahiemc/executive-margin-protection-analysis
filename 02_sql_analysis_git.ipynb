```python
%sql
--Which segments have high revenue but low profit?

SELECT
    customer_segment,
    ROUND(SUM(net_revenue), 2) AS revenue,
    ROUND(SUM(contribution_profit), 2) AS profit,
    ROUND(
        SUM(contribution_profit) /
        NULLIF(SUM(net_revenue), 0) * 100, 2
    ) AS margin_pct,
    ROUND(AVG(discount_pct), 2) AS avg_discount_pct,
    ROUND(AVG(retention_score), 2) AS avg_retention
FROM doordash_margin_orders
GROUP BY customer_segment
ORDER BY revenue DESC;

```

```python
%sql

--Which discount level destroys margin?

SELECT
    CASE
        WHEN discount_pct < 10 THEN '0–10%'
        WHEN discount_pct < 20 THEN '10–20%'
        WHEN discount_pct < 30 THEN '20–30%'
        ELSE '30%+'
    END AS discount_band,
    COUNT(*) AS orders,
    ROUND(AVG(contribution_margin_pct), 2) AS avg_margin_pct,
    ROUND(AVG(retention_score), 2) AS avg_retention,
    ROUND(AVG(refund_rate), 2) AS avg_refund_rate

FROM doordash_margin_orders
GROUP BY discount_band
ORDER BY avg_margin_pct;
```

```python
%sql
--Most profitable customer segment

SELECT
    customer_segment,
    new_vs_repeat,
    subscription_member,
    ROUND(AVG(contribution_profit), 2) AS avg_profit,
    ROUND(AVG(customer_ltv), 2) AS avg_ltv,
    ROUND(AVG(retention_score), 2) AS avg_retention
FROM doordash_margin_orders
GROUP BY 1,2,3
ORDER BY avg_profit DESC;
```

```python
%sql
--Products that need repricing
SELECT
    restaurant_category,
    ROUND(SUM(net_revenue), 2) AS revenue,
    ROUND(AVG(discount_pct), 2) AS avg_discount,
    ROUND(AVG(contribution_margin_pct), 2) AS avg_margin,
    ROUND(AVG(refund_rate), 2) AS avg_refund_rate
FROM doordash_margin_orders
GROUP BY restaurant_category
HAVING AVG(contribution_margin_pct) < 10
ORDER BY revenue DESC;
```

```python
%sql
--Underperforming regions
SELECT
    region,
    ROUND(SUM(net_revenue), 2) AS revenue,
    ROUND(SUM(contribution_profit), 2) AS profit,
    ROUND(
        SUM(contribution_profit) /
        NULLIF(SUM(net_revenue),0) * 100, 2
    ) AS margin_pct,
    ROUND(AVG(refund_rate), 2) AS avg_refund_rate,
    ROUND(AVG(driver_incentive_cost), 2) AS avg_driver_cost
FROM doordash_margin_orders
GROUP BY region
ORDER BY revenue DESC;
```
